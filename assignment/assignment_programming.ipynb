$$
cov(x,y) = \dfrac{(x_1 - \bar{x})(y_1-\bar{y}) + (x_2 - \bar{x})(y_2-\bar{y}) + ... + (x_N - \bar{x})(y_{N}-\bar{y})}{N-1}

= \dfrac{\sum_{i=1}^N (x_i - \bar{x})(y_i - \bar{y})}{N-1}.

The intuition of this quantity is that it looks at the pairs $(x_i, y_i)$ and compares them to the means $(\bar{x},\bar{y})$ to determine whether $x$ and $y$ tend to co-vary in the same direction relative to their means: If the values of $x$ and $y$ are typically both above or below the mean values of $x$ and $y$, then $x$ and $y$ will have a positive covariance, but if $x$ is typically above the mean of $x$ when $y$ is typically below the mean of $y$ or vice versa, then they will have a negative covariance. 6. The **sample correlation coefficient** of two vectors $x$ and $y$ is defined as
$$
r_{xy} = \dfrac{cov(x,y)}{s_x s_y}
$$
Use your functions to create a new function that compute this quantity. The intuition of this quantity is that it is like the covariance, but normalized so that its values like between -1 and 1: perfect negative linear association between the variables at -1, no association at 0, and perfect positive linear association between the variables at 1.

```python
import numpy as np
import math as math
np.random.seed(100) # Set the seed for the random number generator
rho, sigma_x, sigma_y = -.4, 3, 2 # Variance-Covariance Parameters
vcv = np.array([[sigma_x**2, rho*sigma_x*sigma_y],
                [rho*sigma_x*sigma_y,sigma_y**2]]) # VCV Matrix
mu = np.array([-1,2]) # Population averages
sample = np.random.multivariate_normal(mu,vcv,200) # Multivariate normal draws
x = sample[:,0]
y = sample[:,1]

#############################################################################
#1
def sample_mean(x):
    return np.sum(x) / len(x)

x_bar = sample_mean(x)
y_bar = sample_mean(y)

#2
def sample_std(x):
    mean = sample_mean(x)
    return np.sqrt(np.sum((x - mean)**2) / (len(x) - 1))

s_x = sample_std(x)
s_y = sample_std(y)

#3
def z_score(x):
    return (x - sample_mean(x)) / sample_std(x)

z_x = z_score(x)

# 4
def sample_covariance(x, y):
    mean_x = sample_mean(x)
    mean_y = sample_mean(y)
    return np.sum((x - mean_x) * (y - mean_y)) / (len(x) - 1)

cov_xy = sample_covariance(x, y)

# 5
def sample_correlation(x, y):
    return sample_covariance(x, y) / (sample_std(x) * sample_std(y))

corr_xy = sample_correlation(x, y)

#solutions
print("Sample Mean:", x_bar)
print("Sample Standard Deviation:", s_x)
print("Z-scores:", z_x)
print("Covariance:", cov_xy)
print("Correlation:", corr_xy)


```

**Q4.** Optimization is at the core of data science and statistics: Picking the best fit or estimate according to some desirable criteria (e.g. Maximum Likelihood Estimation). In this question, you're going to write a function that finds the highest possible value for some function, and returns the value of the function and the maximizing value.

1. Write a function that creates a grid from some value $a$ to a second value $b$ with $N$ steps, computes the value of a function $f$ on that grid, and then finds the maximizers of the function. Have it return a dictionary with the maximum value and the maximizers themselves.
2. Use your function to maximize $f(x) = 100 - 2x^2 + 3x$ for values of $a=-1$, $b=1$, and for values of $N=3$, $N=10$, $N=100$, $N=1000$ and $N=5000$.
3. The true maximizer of this function is $.75$ and the true maximum value is $101.125$. Is that what you got in the previous step? Why not? How does the quality of the maximization depend on $N$? How does your computed answer change as $N$ gets larger?

```python
import numpy as np
import matplotlib.pyplot as plt

def maximize_on_grid(f, a, b, N):
    x = np.linspace(a, b, N)
    fx = f(x)
    max_value = np.max(fx)
    maximizers = x[fx == max_value]
    return {"maximum": max_value, "maximizer": maximizers}

# Function to plot the grid search and maximizer
def plot_maximization(f, a, b, N):
    x = np.linspace(a, b, N)
    fx = f(x)

    result = maximize_on_grid(f, a, b, N)
    max_x = result["maximizer"]
    max_y = result["maximum"]

    plt.figure()
    plt.scatter(x, fx)
    plt.scatter(max_x, max_y)
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.title(f"Grid Search Maximization (N = {N})")
    plt.show()

    print(result)

# Define the function f(x)
def f(x):
    return 100 - 2*x**2 + 3*x

# Run for required values of N
plot_maximization(f, -1, 1, 3)
plot_maximization(f, -1, 1, 10)
plot_maximization(f, -1, 1, 100)
plot_maximization(f, -1, 1, 1000)
plot_maximization(f, -1, 1, 5000)
```

3.The computed maximizer does not exactly equal the true maximizer of 0.75 for small values of N because the grid only evaluates the function at a finite number of points. As N increases, the grid becomes finer, making it more likely that a value close to 0.75 is included. As a result, both the estimated maximizer and maximum value converge toward the true values as N gets larger.
